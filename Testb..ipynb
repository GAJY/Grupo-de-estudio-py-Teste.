    El activo AMZN tiene la mayor varianza de todod el dataset con ______ y una media de _____, sesgo ______ y desviación estándar ______. 

Recurda que la forma de calcular los rendimientos de una serie de tiempo es: 

$$\frac{Y_{t}}{Y_{t-1}}-1$$

Donde $Y_{t}$ es el precio actual y $Y_{t-1}$ es el precio anterior. 

```python
# Punto 1 Actividad 2. 
df_ren=pd.DataFrame()
for activo in df:
    df_ren[f'rendimiento de:{activo}']=np.log(df[activo].shift()/df[activo])
df_ren.head()
```

Para hacer el calculo de los rendimientos logaritmicos de cada una de las series (columnas) de la base de datos creada en el punto 2 de la actividad 1 se ocuparán dos funciones np.log() y .shift().
- Lo que hace la función shift es crear un resago en las series. 
- Lo que hace la función np.log() (de la paquetería de numpy) es combertir a logaritmo cualquier valor que se le introdusca dentro de los paréntesis. 

Como de nada sirve hacer el calculo sin tener la información de una manera accesible, se debe guardar en algún lado.

Por esto para almacenar la información se crea un dataframe basio usando la paquería de pandas con la funcion dataframe (pd.DataFrame) y este dataframe basio se le atribuye a la variable df_ren que será la que contendrá el dataframe con los calculos realizados. 

Ahora: 

Como se debe de llenar el dataframe creado con los calculos y estos se hacer serie por serie (columna por columna), se hace uso de un ciclo for como el anteriormente visto. 

Por lo tanto, a cada activo -columna- dentro del df (for activo in df), lo resagará (df[activo].shift) y lo dividirá por ese mismo activo no resagado (df[activo].shift()/df[activo]). Y como esto se encuentra dentro de la función de numpy log, a dicha operación se le sacará el logaritmo. Además toda esta operación se orienta (=) a (df_ren[f'rendimiento de:{activo}']) que la forma en la cual a cada operación hecha para cada activo se le creará una columna dentro del nuevo dataframe. 

Dicho de otra forma: 

Se crea una nueva columna para el dataframe creado, que contendrá la operación hecha y esto se hará para cada uno de los activos dentro de df.
